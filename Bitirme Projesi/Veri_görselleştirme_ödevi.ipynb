# 50 Startups Veri Analizi
Bu Notebook, `50_Startups.csv` dosyasındaki startup şirketlerinin harcamalarını ve kârlılıklarını analiz etmek için hazırlanmıştır.

Amaç: Ar-Ge, yönetim ve pazarlama harcamalarının kâr üzerindeki etkisini incelemek ve görselleştirmektir.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
get_ipython().run_line_magic('matplotlib', 'inline')
data = pd.read_csv('50_Startups.csv')
data.head()
```

## Görev 1: R&D harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterin.

```python
plt.figure(figsize=(8,5))
plt.scatter(data['R&D Spend'], data['Profit'], color='blue', alpha=0.9)
plt.title("R&D Harcaması ile Kâr İlişkisi")
plt.xlabel("R&D Harcaması")
plt.ylabel("Kar")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Noktaların soldan sağa doğru yukarı gitmesi doğrusal pozitif ilişki olduğunu gösteriyor. Yani Ar-Ge’ye yapılan yatırım ile kâr arasında neredeyse doğru orantı var.
Sonuç olarak Ar-Ge harcaması arttıkça kâr da artıyor.



## Görev 2: Yönetim harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterin.


```python
plt.figure(figsize=(8,5))
plt.scatter(data['Administration'], data['Profit'], color='red', alpha=0.9)
plt.title("Yönetim Harcamaları ve Kâr Arasındaki İlişki")
plt.xlabel("Yönetim Harcaması")
plt.ylabel("Kar")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Noktaların dağınık durumda olması, yönetim harcamaları ile kâr arasında belirli bir ilişki olmadığını gösterir.
Sonuç olarak, yönetim harcamalarının Ar-Ge harcamalarına göre kârı açıklama gücü çok daha düşük.

## Görev 3: Eyaletlere göre ortalama kârları bar chart ile karşılaştırın.


```python
state_means = data.groupby('State')['Profit'].mean() #verideki her eyaletin ortalama kârını hesapladım.
plt.figure(figsize=(8,5))
state_means.plot(kind='bar', color=['skyblue','pink','purple'])
plt.title("Eyaletlere Göre Ortalama Kâr")
plt.xlabel("Eyalet")
plt.ylabel("Ortalama Kar")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Florida'nın en yüksek ortalama kâra sahip olduğunu gözlemliyoruz.
New York ikinci sırada, California ise üçüncü sırada.

## Görev 4: R&D, yönetim ve pazarlama harcamalarının dağılımını boxplot ile karşılaştırın.

```python
plt.figure(figsize=(8,5))
plt.boxplot(
    [data['R&D Spend'], data['Administration'], data['Marketing Spend']],
    labels=['R&D Spend', 'Administration', 'Marketing Spend'])
plt.title("Harcama Türlerinin Karşılaştırması")
plt.xlabel("Yönetim Harcaması")
plt.ylabel("Harcama Miktarı")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Kutu (box): ortadaki %50’lik veri aralığı (Q1–Q3)

Ortadaki çizgi: medyan (ortalama değil, tam ortadaki değer)

Kuyruklar (whiskers): minimum ve maksimum değerler (uç değerler hariç)

Olası noktalar (dotlar varsa): uç değerler (aykırı veriler)

1. R&D Spend : Medyan düşük, dağılım az, harcamalar birbirine yakın.
2. Administration (Yönetim Harcaması): Medyan yaklaşık 120.000, yönetim harcamaları birbirine yakın ve stabil.
3. Marketing Spend : Dağılım geniş, medyan yaklaşık 200.000 olmasına rağmen maksimum değer 450.000'e kadar çıkmış. Yani pazarlamada harcama farkı çok yüksek.

Sonuç olarak ;
Pazarlama harcamaları en değişken kalem.

Yönetim harcamaları daha dengeli ve sabit.

Ar-Ge harcamaları düşük ama istikrarlı bir aralıkta.
