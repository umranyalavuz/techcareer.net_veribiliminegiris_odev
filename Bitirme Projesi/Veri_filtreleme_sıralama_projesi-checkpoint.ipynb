# Ülke Verileri Analizi
Bu Notebook, `country.csv` dosyasındaki ülke verilerini analiz etmek için hazırlanmıştır.
Her görevde farklı bir kriter üzerinden sıralamalar ve filtrelemeler yapılmıştır.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('country.csv')
data
```

## Görev 1: Nüfusa Göre Azalan Sıralama
- Amaç: En kalabalık 10 ülkeyi belirlemek ve görselleştirmek.
- Yöntem: `Population` sütununa göre azalan sıralama yapılmıştır.



```python
pop_sorted = data.sort_values(by="Population", ascending=False).head(10) ##sort_values ile azalan sırayla sıralama yaptım. Bu satırdaki amaç en kalabalık 10 ülkeyi bulmak.
print("\nNüfusa göre ilk 10 ülke:\n", pop_sorted[["Country","Population"]])
plt.figure(figsize=(8,4))
sns.barplot(x="Population", y="Country", data=pop_sorted, palette="magma" )
plt.title("Nüfusa Göre İlk 10 Ülke")
plt.show()
```

En kalabalık ülke: Çin

Sonrasında Hindistan

Sonrasında ABD, Endonezya, Brezilya...

Listenin sonunda Japonya yer alıyıor.

## Görev 2: GDP per capita Artan Sıralama
- Amaç: Kişi başına düşen geliri en düşük 10 ülkeyi bulmak ve görselleştirmek.
- Yöntem: `GDP ($ per capita)` sütununa göre artan sıralama yapılmıştır.


```python
gdp_sorted = data.sort_values(by="GDP ($ per capita)", ascending=True).head(10) ##ascending=True ile artan sıralama yaptım. Bu satırdaki amaç kişi başına gelir en düşük 10 ülkeyi bulmak.
print("\nGDP per capita en düşük 10 ülke:\n", gdp_sorted[["Country","GDP ($ per capita)"]])
    
plt.figure(figsize=(8,4))
sns.barplot(x="GDP ($ per capita)", y="Country", data=gdp_sorted, palette="magma" )
plt.title("GDP per capita En Düşük 10 Ülke")
plt.show()
```

En düşük kişi başına gelir sırasıyla: Doğu Timor, Sierra Leone ve Somali.

Sonrasında Burundi, Malawi, Tanzanya ve Gazze Şeridi geliyor.

Listenin sonunda ise Komorlar, Kongo Demokratik Cumhuriyeti ve Etiyopya yer alıyor.

Genel olarak tüm ülkeler 500–700 $ aralığında ve küresel ortalamanın oldukça altında.

## Görev 3: Nüfusu 10 Milyon Üzeri Ülkeler
- Amaç: Nüfusu 10 milyonun üzerinde olan ülkeleri seçmek ve ilk 10 ülkeyi görselleştirmek.


```python
over_10m = data[data["Population"] > 10_000_000]
print("\nNüfusu 10 Milyondan Fazla Ülkeler", over_10m[["Country","Population"]].head(10))
over_10m_top10 = over_10m.sort_values(by="Population", ascending=False).head(10)
plt.figure(figsize=(8,4))
sns.barplot(x="Population", y="Country", data=over_10m_top10, palette="pastel")
plt.title("Nüfusu 10 Milyondan fazla İlk 10 Ülke")
plt.show()
```

Dünya nüfusunun büyük kısmı, Çin ve Hindistan başta olmak üzere bu 10 ülke içinde yoğunlaşmıştır.

## Görev 4: En Yüksek Okur-Yazarlık Oranı
- Amaç: Literacy (%) sütununa göre sıralama yaparak en yüksek okuryazarlık oranına sahip ilk 5 ülkeyi bulmak.


```python
top5_literacy = data.sort_values(by="Literacy (%)", ascending=False).head(5)
print("\nEn Yüksek Okur-Yazarlık oranına Sahip 5 Ülke", top5_literacy[["Country","Literacy (%)",]])
plt.figure(figsize=(8,4))
sns.barplot(x="Literacy (%)", y="Country", data=top5_literacy, palette="Blues_r")
plt.title("En Yüksek Okuryazarlık oranına Sahip 5 Ülke")
plt.show
```

Bu ülkelerde okuryazarlık oranı neredeyse %100’e ulaştığı için, eğitim sistemlerinin güçlü, erişilebilir ve toplumun geneline yaygın olduğu görülüyor.

En yüksek okuryazarlık oranına sahip ülke: Çekya (Czech Republic)

Sonrasında: İzlanda

Ardından: Polonya

Devamında: Letonya

Listenin sonunda: Estonya yer alıyor.

## Görev 5: Kişi Başı GSYİH 10.000'in Üzerinde Olan Ülkeler
- Amaç: GDP ($ per capita) sütunu 10.000'in üzerinde olan ülkeleri seçmek ve ilk 10 ülkeyi görselleştirmek.


```python
gdp_over_10000= data[data["GDP ($ per capita)"] > 10000]
gdp_over_10000_top10 = gdp_over_10000.sort_values(by="GDP ($ per capita)", ascending=False).head(10)
print(gdp_over_10000[["Country", "GDP ($ per capita)"]])
plt.figure(figsize=(8,4))
sns.barplot(x="GDP ($ per capita)", y="Country", data=gdp_over_10000_top10, palette="coolwarm")
plt.title("Kişi Başı GSYİH 10000 $ üzerinde olan ülkeler (ilk 10)")
plt.show()
```

En yüksek kişi başı GSYİH'ya sahip ülke Lüksemburg’dur.

Onu sırasıyla Norveç, ABD, Bermuda, Cayman Adaları ve diğer gelişmiş ekonomiler takip etmektedir.

## Görev 6: En Yüksek Nüfus Yoğunluğuna Sahip İlk 10 Ülke
- Amaç: Pop. Density (per sq. mi.) sütununa göre sıralama yapıp en yüksek nüfus yoğunluğuna sahip ilk 10 ülkeyi bulmak ve görselleştirmek.


```python
top10_density = data.sort_values(by="Pop. Density (per sq. mi.)", ascending=False).head(10)
print("\nEn Yüksek Nüfus Yoğunluğuna Sahip İlk 10 Ülke:\n", top10_density[["Country", "Pop. Density (per sq. mi.)"]])
plt.figure(figsize=(8,4))
sns.barplot(x="Pop. Density (per sq. mi.)", y="Country", data=top10_density,palette="coolwarm")
plt.title("En Yüksek Nüfus Yoğunluğuna Sahip 10 Ülke")
plt.show
```

X ekseni: Nüfus yoğunluğu (Pop. Density per sq. mi.)

Y ekseni: Ülke isimleri (Country)

Ermenistan (Armenia) grafikte en yüksek çubukla temsil edilmiş; nüfus yoğunluğu diğerlerine göre az farkla da olsa en yüksek.

Türkiye ise bu sıralamada 10. sırada, diğer ülkelere göre daha düşük yoğunluklu görünüyor.
