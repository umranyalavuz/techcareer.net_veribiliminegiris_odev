# Dava Verisi Kümeleme Analizi

Bu notebook’ta bir dava veri setini kullanarak K-Means kümeleme analizi yapacağız.  
Veri setindeki özellikler ve görevler aşağıda belirtilmiştir:

**Veri Seti Özellikleri:**  
- Case Duration (Gün): Davanın tamamlanması için geçen süre (gün cinsinden).  
- Number of Witnesses (Tanık Sayısı): Dava boyunca dinlenen tanık sayısı.  
- Legal Fees (Hukuk Maliyetleri): Dava süresince oluşan toplam hukuk maliyetleri (USD cinsinden).  
- Number of Evidence Items (Delil Sayısı): Davada kullanılan delil sayısı.  
- Severity (Ciddiyet Düzeyi): 1: Düşük, 2: Orta, 3: Yüksek.  
- Outcome (Sonuç): 0: Aleyhte, 1: Lehinde.

**Görevler:**  
1. Özellik seçimi  
2. Küme sayısını belirleme (Elbow yöntemi)  
3. K-Means ile kümeleme  
4. Sonuçları görselleştirme ve yorumlama

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
# Veri setini okuma
data = pd.read_csv('dava.csv')
data.head()
```

# Kümeleme için uygun özellikleri seçin. Elbow yöntemi ile optimal küme sayısını belirleyin.
- Case Duration (Days)  
- Number of Witnesses  
- Legal Fees (USD)  
- Number of Evidence Items

```python
features = data[["Case Duration (Days)", "Number of Witnesses", "Legal Fees (USD)", "Number of Evidence Items" ]]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

inertia = []
K = range(1, 11)
for k in K :
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, "bo-")
plt.xlabel("Küme Sayısı (k)")
plt.ylabel("Intertia")
plt.title("Elbow Yöntemi")
plt.show()    
```

Grafikte Küme Sayısı (k) arttıkça inertia değeri düşüyor. Ancak eğri bir noktadan sonra yavaşlamaya başlıyor, bu yavaşlamanın başladığı kısısm  dirsek noktası oluyor. Bu da en uygun küme sayısını (k) gösteriyor.

# K-Means algoritması ile veriyi kümeleyin. Sonuçları grafiklerle görselleştirip yorumlayın

Elbow yönteminden belirlenen küme sayısı k=3 olarak seçildi.  
K-Means algoritması ile kümeleme yapıp sonuçları görselleştiriyoruz.

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
data["Cluster"] = kmeans.fit_predict(scaled_features)

# Görselleştirme
sns.scatterplot(x="Case Duration (Days)", y="Legal Fees (USD)", 
                hue="Cluster", data=data, palette="viridis")
plt.title("K-Means Kümeleme (k=3)")
plt.show()
print("Küme Merkezleri (ölçeklenmiş):")
print(kmeans.cluster_centers_)
```

Bu grafikte K-Means algoritması, davaları üç farklı gruba ayırmış:
1. kısa ve ucuz,

2. orta düzeyde,

   

3. uzun ve pahalı olanlar.

Merkezler ise bize her kümenin genel eğilimini gösteriyor:

1. merkez: yüksek ücret, kısa süre

2. merkez: düşük ücret, kısa süre

3. merkez: yüksek süre, orta ücret

Ayrıca üç farklı renkli grup var:

1. Cluster 0 (mor tonları): Genellikle orta süreli davalar ve orta-yüksek ücretler.

2. Cluster 1 (yeşilimsi tonlar): Daha kısa süreli ve düşük ücretli davalar.

3. Cluster 2 (sarı noktalar): Uzun süren ve yüksek ücretli davalar.
